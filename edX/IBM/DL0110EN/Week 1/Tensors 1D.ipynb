```python
import torch
```

```python
a = torch.tensor([0, 1, 2, 3, 4])
print(a.dtype)
print(a.type) # torch.LongTensor
```

```python
a = torch.tensor ([0.0, 1.0, 2.0, 3.0, 4.0]) 
print(a.dtype)
print(a.type) # torch.FloatTensor
```

### Casting tensor

```python
a = torch.FloatTensor([0, 1, 2, 3, 4])
print(a.dtype)
print(a.type) # torch.FloatTensor
# or 
a = a.type(torch.LongTensor)
print(a.dtype) # torch.LongTensor
```

### Checking dimensions

```python
print(a.size())
print(a.ndimension())
```

### Converting shape

```python
a_col = a.view(5, 1)
print(a_col)

# converting to a column vector
a_cv = a.view(-1, 1)
print(a_cv)
```

### Converting from numpy and to numpy

```python
import numpy as np

a = np.array([0, 1, 2, 3, 4, 5])
b = torch.from_numpy(a)
print(b)

c = b.numpy()
c
```

### Converting from pandas

```python
import pandas

pds = pandas.Series([0.0, 1.0, 2.0])
a = torch.from_numpy(pds.values)
a
```

## Indexing and Slicing

```python
c = torch.tensor([20, 1, 2, 3, 4])
print(c[0])
c[0] = 100
print(c[0])
```

```python
# slicing
d = c[1:4]
d
```

```python
c[3:5] = torch.tensor([100, 200])
c
```

## Vector Operations

```python
# Adition
a = torch.tensor([1, 2, 3, 4])
b = torch.tensor([10, 20, 30, 40])
c = a+b
c
```

```python
# multiply with a scalar
c = 3*a
c
```

```python
# Element-wise multiplication
c = a*b
c
```

```python
# Dot product
c = torch.dot(a, b)
c
```

```python
# Broadcasting (adding a number to the tensor)
c = 1+a
c
```

### Universal Operations

```python
# Calculate the Mean
a = torch.tensor([1., 2., 3., 4.])
mean_a = a.mean()
mean_a
```

```python
# Get the maximum value
max_a = a.max()
max_a
```

```python
# Create a vector  with values
b = torch.linspace(-2, 2, 5)
b
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

a = torch.linspace(-10, 10, 100)
sina = torch.sin(a)
plt.plot(a.numpy(), sina.numpy())
plt.show()
```
